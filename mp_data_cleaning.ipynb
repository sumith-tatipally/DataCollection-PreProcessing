# Email

Variable Email contains element like [DOT], [AT] that clearly means "." or "@" in email format context. We will now try to bring them in correct format

```python
mp_data["email"]= mp_data["email"].replace(to_replace ="[[DOT][DOT][DOT][DOT]]", value =".", regex=True)
mp_data["email"]= mp_data["email"].replace(to_replace ="[[[AT][AT][AT]]", value ="@", regex=True)
mp_data["email"]= mp_data["email"].replace(to_replace =" ", value =",", regex=True)
mp_data.head()
```

We started replace function by assuming [AT] and [DOT] as simple string but the technique didnot work beacause these are characters that appeared like these due web element scrapping. We had to take help of "https://regexr.com/" to find right regex to replace character and bring them in proper format.   

Our data has email ID missing for a lot of MP's but that's understandable because email ID's have became a prevalent method of communication last decade or so. Hence, we are ok with email variable and move to next

# State

Apart from 33 missing values in variable State, we can observe that eventhough the variable name says called "State". It contains constituency information. Let's separate state from consituency and then handle missing ones. Here, we cannot simply use split by character because it has (SC or ST) in between.

SC and ST in state signifies if if the constiuency is under reservation quota. Hence, We created a column that depicts if constituency is under reservation(1) or not(0)

```python
mp_data["Reservation"]= pd.np.where(mp_data.State.str.contains("SC"),"Yes",
                                   pd.np.where(mp_data.State.str.contains("ST"),"Yes","No"))

mp_data.head()                                    
```

Now, We need to split with delimeter and then remove special characters

```python
mp_data[['Constituency', 'State']]= mp_data['State'].str.split('(', n=1, expand=True)
```

```python
mp_data["State"]= mp_data["State"].replace(to_replace ="SC\W\W", value ="",  regex=True)
mp_data["State"]= mp_data["State"].replace(to_replace ="[!,*)@#%(&$_?.^]", value ="",  regex=True)
mp_data.head()
```

We now have a separate column for Constituency and State that will make data analysis easy

# DOB
1. DOB has Many missing values, we scrapped data again using different method and will use the output to fill missing data in earlier scrapped data
 

```python
mp11term= pd.read_csv('mp11term.csv')
mp12term= pd.read_csv('mp12term.csv')
mp_data["DOB"].isna().sum()
```

```python
mp_data=mp_data.merge(mp11term, on='Name', how='left')
mp_data=mp_data.merge(mp12term, on='Name', how='left')
mp_data.columns
```

```python
#Removing Unwanted columns
mp_data.drop("Education11", axis=1, inplace=True)
mp_data.drop("Education12", axis=1, inplace=True)
mp_data.drop("Profession11", axis=1, inplace=True)
mp_data.drop("Profession12", axis=1, inplace=True)
mp_data.drop("Permanent_Address11", axis=1, inplace=True)
mp_data.drop("Permanent_Address12", axis=1, inplace=True)
mp_data.drop("Present_Address11", axis=1, inplace=True)
mp_data.drop("Present_Address12", axis=1, inplace=True)
mp_data.drop("Unnamed: 0", axis=1, inplace=True)
mp_data.drop("Unnamed: 0_y", axis=1, inplace=True)
mp_data.drop("Unnamed: 0_x", axis=1, inplace=True)
mp_data.columns
```

```python
mp_data["DOB"]=mp_data["DOB"].fillna(mp_data["DOB11"])
mp_data["DOB"]=mp_data["DOB"].fillna(mp_data["DOB12"])
mp_data.drop("DOB11", axis=1, inplace=True)
mp_data.drop("DOB12", axis=1, inplace=True)
mp_data["DOB"].isna().sum()
```

We have managed to reduce more than 200 missing values with this method

#### 2. Calculation Age as on today from date of birth variable

```python
mp_data["DOB"]= pd.to_datetime(mp_data["DOB"], format="%d/%m/%Y")

def from_dob_to_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

mp_data["Age"]=mp_data["DOB"].apply(lambda x: from_dob_to_age(x))
```
