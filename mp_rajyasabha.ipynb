```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import Select
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import json
import time
from collections import Counter
from bs4 import BeautifulSoup
from collections import Counter
```

```python
# chrome driver to run selenium
driver = webdriver.Chrome(executable_path='D:\\ISB\\chromedriver_win32\\\\chromedriver.exe')


# seed url to get the data
URL = "https://rajyasabha.nic.in/Members/AlphabeticalFormerMember"

driver.get(URL)


# To wait for the elements to be visible / rendered by the browser
# wait = WebDriverWait(driver, 10)
time.sleep(3)
```

```python
#we have a tabular data and all the details are in alphabetical order. so we are trying to get hyperlinks for all the alphabets
# separately and appending them to a list.


soup1 = BeautifulSoup(driver.page_source,"html.parser") #parse the url to get all the alphabets

alpha_letters = soup1.find('div',{'id':'English'}) 

all_alpha_let = alpha_letters.find_all('a') 
print(len(all_alpha_let))

res = []
for each in all_alpha_let:
    # print(each.text)
    alp_l = each.text
    driver.find_element_by_link_text(alp_l).click() #click on each alpha (loop)
    time.sleep(3)
    # To wait for the elements to be visible / rendered by the browser
    soup2 = BeautifulSoup(driver.page_source,"html.parser")
    
    all_alpha = soup2.find('div', {'class': 'searchResulltTabl'}) #fetch members table for each alphabet
    all_aaa = all_alpha.find('tbody',{'id':'memberList'})
    all_tr = all_aaa.find_all('tr')
    
    for ea in all_tr:
        # print(ea)
        all_td = ea.find_all('td')
        
        name = all_td[2].text
        print(name)
        
        time.sleep(2) # To wait for the elements to be visible / rendered by the browser
        results = dict()
        
        if len(name) > 2:

            time.sleep(3)
            driver.find_element_by_link_text(name).click() # click on the name to get the pop up window to fetch more information
            time.sleep(3)
            # print(each)
            
            soup4 = BeautifulSoup(driver.page_source,"html.parser") #parse the pop up window
            time.sleep(3) ## To wait for the elements to be visible / rendered by the browser
            bbb = soup4.find('table',{'id':'biodata_Table'}) #fetch the table and find all the rows in it using td tag
            
            all_tttr = bbb.find_all('td')
            
            try:
                father = all_tttr[1].text.strip() # extract fathername from popup window
            except:
                father = 'NA'

            try:
                mother = all_tttr[2].text.strip() # extract mother name from popup window
            except:
                mother = 'NA'

            try:
                dob = all_tttr[3].text.strip() # extract date of birth from popup window
            except:
                dob = 'NA'

            try:
                pob = all_tttr[4].text.strip() # extract place of birth from popup window
            except:
                pob = 'NA'

            try:
                marsta = all_tttr[5].text.strip() # extract marriage status from popup window
            except:
                marsta = 'NA'

            try:
                wifen = all_tttr[6].text.strip() # extract wife name from pop up window
            except:
                wifen = 'NA'

            try:
                educ = all_tttr[8].text.strip() # extract education from pop up window
            except:
                educ = 'NA'

            try:
                profesn = all_tttr[9].text.strip() # extract profession from pop up window
            except:
                profesn = 'NA'

            try:
                addr = all_tttr[10].text.strip() #extract address from popup window
            except:
                addr = 'NA'

            try:
                emai = all_tttr[12].text.strip() #extract email from popup window
            except:
                emai = 'NA'
                
            time.sleep(3) ## To wait for the elements to be visible / rendered by the browser
            driver.find_element_by_class_name('close').click() #clicking on close button of pop window
            
            
        try:
            sno = all_td[0].text.strip() #serial no
        except:
            sno='NA'

        try:
            name = all_td[2].text.strip() #candidate name
        except:
            name='NA'

        try:
            state =  all_td[3].text.strip() # state name
        except:
            state = 'NA'

        try:
            party = all_td[4].text.strip() #party name
        except:
            party = 'NA'

        try:
            term = all_td[5].text.strip() #term period
        except:
            term = 'NA'


        try:
            vaca = all_td[6].text.strip() #vacation data/reason
        except:
            vaca = 'NA'


        results['Sno'] = sno
        results['Name'] = name
        results['State'] = state
        results['party'] = party
        results['term'] = term
        results['vacation'] = vaca
        results['father_name'] = father
        results['mother_name'] = mother
        results['dob'] = dob
        results['pob'] = pob
        results['marriage_status'] = marsta
        results['wife_name'] = wifen
        results['education'] = educ
        results['Profession'] = profesn
        results['address'] = addr
        results['email'] = emai
        res.append(results)

    
```

```python
with open('mp_rajya.json', 'w') as f:  # dumping values to json
    json.dump(res, f)
```
